```python
import pandas as pd
import json
wbp_df = pd.read_json('world_bank_projects.json')
```

```python
# 1. Find the 10 countries with most projects
# Create smaller dataframe with only a column for the country name and another for the relevant count.
country_entries = wbp_df[['countryname', 'project_name']]

# Organize the dataframe to explore the results. 
country_entries = country_entries.groupby('countryname')

# Count the number of entries per country, sort them in decreasing order, print the top ten countries.
country_entries = country_entries.size().sort_values(ascending=False)[:10]
print ("10 countries with the most projects: \n\n", country_entries)
```
